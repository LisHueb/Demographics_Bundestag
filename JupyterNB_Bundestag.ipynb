# extract relevant information from XML file

```python
import xml.etree.ElementTree as ET
import pandas as pd

# read XML
tree = ET.parse('MDB_STAMMDATEN.XML')
root = tree.getroot()

data = []

# Parse MDB Elements
for mdb in root.findall('MDB'):
    # personal information
    person_data = {
        'id': mdb.find('.//ID').text,
        'last_name': mdb.find('.//NACHNAME').text,
        'first_name': mdb.find('.//VORNAME').text,
        'adel': mdb.find('.//ADEL').text,
        'prefix': mdb.find('.//PRAEFIX').text, 
        'title': mdb.find('.//ANREDE_TITEL').text,
        'academic_title': mdb.find('.//AKAD_TITEL').text,
        'birthday': mdb.find('.//GEBURTSDATUM').text,
        'place_of_birth': mdb.find('.//GEBURTSORT').text,
        'country_of_birth': mdb.find('.//GEBURTSLAND').text,
        'date_of_death': mdb.find('.//STERBEDATUM').text,
        'gender': mdb.find('.//GESCHLECHT').text,
        'family_status': mdb.find('.//FAMILIENSTAND').text,
        'religion': mdb.find('.//RELIGION').text,
        'profession': mdb.find('.//BERUF').text,
        'party': mdb.find('.//PARTEI_KURZ').text      
        
    }
    
    # information regarding election period for each person
    for periode in mdb.findall('.//WAHLPERIODE'):
        wp_data = person_data.copy()
        wp_data['election_period'] = periode.find('WP').text
        wp_data['mdb_from'] = periode.find('MDBWP_VON').text
        wp_data['mdb_until'] = periode.find('MDBWP_BIS').text  
        wp_data['electoral_district_nr'] = periode.find('WKR_NUMMER').text
        wp_data['electoral_district_name'] = periode.find('WKR_NAME').text
        wp_data['electoral_district_state'] = periode.find('WKR_LAND').text
      
        data.append(wp_data)

# create dataframe
df = pd.DataFrame(data)

df
```

# data wrangling

```python
df.info()
```

```python
# change data types where necessary

As_datetime = ['birthday', 'date_of_death', 'mdb_from', 'mdb_until']
df[As_datetime] = df[As_datetime].apply(pd.to_datetime)

df['election_period'] = df['election_period'].astype(int)
df['id'] = df['id'].astype(int)

df.info()
```

```python
#the information regarding the electoral districts seems incomplete --> delete these columns because I cannot use them anyway

df= df.drop(columns=['electoral_district_nr', 'electoral_district_name', 'electoral_district_state'])
```

```python
# mdb_until (end of electoral period) has less values than mdb_from (beginning electoral period)
# --> confirm the assumption that these missing values belong to the current election period because it does not have an end date yet 
df2 = df[df['mdb_until'].isnull()]
df2['election_period'].unique()
```

```python
## country_of_birth does not have values for every person. Assumption: country is not specified if it's Germany 

df3 = df[df['country_of_birth'].isnull()]
df3

df3 = df[df['country_of_birth']=="Deutschland"]
df3

# assumption confirmed ("Deutschland" only added from period 19)
```

```python
df['country_of_birth'] = df['country_of_birth'].fillna('Deutschland')
```

```python
# I'm interested in the family status but it seems very versatile and therby hard to normalize
df['family_status'].unique() 
```

```python
# I decide to extract the number of children instead because that seems easier to normalize

import re
def extract_number_of_children(status):
    if pd.notna(status):
        # I use *kind.*" because I also have values like Adoptivkinder or Pflegekinder which I want to include
        match = re.search(r'(\d+)\s*.*kind.*', status, flags=re.IGNORECASE)
        if match and match.group(1) is not None:
            return int(match.group(1))
    return 0  


df['number_of_children'] = df['family_status'].apply(extract_number_of_children)
df['number_of_children'] = df['number_of_children'].astype(int)

print(df[['family_status', 'number_of_children']])
```

```python
# check that each person has only one party (many have more than one election period and could possibly change party between two periods)
party_per_person = df.groupby(['id'])['party'].nunique().to_frame('nr_parties').reset_index()
party_per_person[party_per_person['nr_parties']>1]
```

```python
# check religion values (df.info showed some missing values for religion)
check_rel= df[['id','religion']].drop_duplicates(subset = 'id')

rel_ohneAngabe= check_rel[check_rel['religion'].isin([None, 'ohne Angaben'])].shape[0]
print('Person Ids without information:', rel_ohneAngabe)

rel_mitAngabe= check_rel[~check_rel['religion'].isin([None, 'ohne Angaben'])].shape[0]
print('Person Ids with information:', rel_mitAngabe)

#--> religion not usable for analysis --> drop column
df= df.drop(columns=['religion'])
```

```python

engine = create_engine(f'postgresql+psycopg2://{user}:{password}@{host}:{port}/{dbname}')

# adjust data types
sql_datatypes = {
    'object': 'TEXT',
    'datetime64[ns]': 'TIMESTAMP', 
    'int32': 'INTEGER'
}


# create table 
with engine.connect() as conn:
    # iterating over the columns of df and concatenating the column name with corresponding data type (from sql_datatypes) 
    columns_info = ', '.join([f"{col} {sql_datatypes[str(df[col].dtype)]}" for col in df.columns])
    conn.execute(f'''
    CREATE TABLE IF NOT EXISTS rawdata (
        {columns_info}
    );
    ''')

# insert data
df.to_sql('rawdata', engine, if_exists='replace', index=False)

conn.close()
```
